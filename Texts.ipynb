                                        1. TEXT ENCODING/DECODING

```python
#Sort node information by nodeid
all_files = os.listdir("node_information/text/")
sortedFiles = sorted(all_files,key=lambda x: int(os.path.splitext(x)[0]))
results = defaultdict(list)

#For each node information, find its encoding, and decode it. Allows for accents to be 
#properly decoded. This whole process is really important and allows for a proper words
#retrieval. However, this process might take several hours (3-5). (Need to be done once)

count = 0
for file in sortedFiles:
    path = 'node_information/text/'+ file
    # Detect encoding
    with open(path, 'rb') as ef:
        encoding =detect(ef.read())['encoding']
        print(count,encoding)
        if encoding != "utf-8" and encoding != "ascii" :
            encoding = 'Latin-1'
        #Try the encoding found
        try:
            with open(path,'r', encoding = encoding) as file_open:
                results["text"].append(file_open.read())
        except:
            with open(path,'r', encoding='Latin-1') as file_open:
                results["text"].append(file_open.read())
        count=count+1

# Put all decoded node informations in a dataframe.        
textDf = pd.DataFrame(results)           
    
```

```python
"""                                                                                  
The major problem encountered with this challenge is the presence of both english and french languages. 
Because using only english or french tokenizer/lemmatizer will yield bad results, the two are used. 
To decide which one to use, a language decoder is used giving language proportions in texts.
For the french lemmatization, scapy was used and for the english one, gensim/nltk. 
Same as before, this process takes at least one hour and has been used to get an almost perfect lemmatization.
"""

#French stopwords and the "alternate" word found in many documents.
FRENCH_STOPWORDS = ["alternate","a","abord","absolument","afin","ah","ai","aie","ailleurs","ainsi","ait","allaient","allo","allons","all","alors","anterieur","anterieure","anterieures","apres","aprs","as","assez","attendu","au","aucun","aucune","aujourd","aujourd'hui","aupres","auquel","aura","auraient","aurait","auront","aussi","autre","autrefois","autrement","autres","autrui","aux","auxquelles","auxquels","avaient","avais","avait","avant","avec","avoir","avons","ayant","b","bah","bas","basee","bat","beau","beaucoup","bien","bigre","boum","bravo","brrr","c","car","ce","ceci","cela","celle","celle-ci","celle-l","celles","celles-ci","celles-l","celui","celui-ci","celui-l","cent","cependant","certain","certaine","certaines","certains","certes","ces","cet","cette","ceux","ceux-ci","ceux-l","chacun","chacune","chaque","cher","chers","chez","chiche","chut","chre","chres","ci","cinq","cinquantaine","cinquante","cinquantime","cinquime","clac","clic","combien","comme","comment","comparable","comparables","compris","concernant","contre","couic","crac","d","da","dans","de","debout","dedans","dehors","deja","del","depuis","dernier","derniere","derriere","derrire","des","desormais","desquelles","desquels","dessous","dessus","deux","deuxime","deuximement","devant","devers","devra","different","differentes","differents","diffrent","diffrente","diffrentes","diffrents","dire","directe","directement","dit","dite","dits","divers","diverse","diverses","dix","dix-huit","dix-neuf","dix-sept","dixime","doit","doivent","donc","dont","douze","douzime","dring","du","duquel","durant","ds","dsormais","e","effet","egale","egalement","egales","eh","elle","elle-mme","elles","elles-mmes","en","encore","enfin","entre","envers","environ","es","est","et","etant","etc","etre","eu","euh","eux","eux-mmes","exactement","except","extenso","exterieur","f","fais","faisaient","faisant","fait","faon","feront","fi","flac","floc","font","g","gens","h","ha","hein","hem","hep","hi","ho","hol","hop","hormis","hors","hou","houp","hue","hui","huit","huitime","hum","hurrah","h","hlas","i","il","ils","importe","j","je","jusqu","jusque","juste","k","l","la","laisser","laquelle","las","le","lequel","les","lesquelles","lesquels","leur","leurs","longtemps","lors","lorsque","lui","lui-meme","lui-mme","l","ls","m","ma","maint","maintenant","mais","malgre","malgr","maximale","me","meme","memes","merci","mes","mien","mienne","miennes","miens","mille","mince","minimale","moi","moi-meme","moi-mme","moindres","moins","mon","moyennant","multiple","multiples","mme","mmes","n","na","naturel","naturelle","naturelles","ne","neanmoins","necessaire","necessairement","neuf","neuvime","ni","nombreuses","nombreux","non","nos","notamment","notre","nous","nous-mmes","nouveau","nul","nanmoins","ntre","ntres","o","oh","oh","oll","ol","on","ont","onze","onzime","ore","ou","ouf","ouias","oust","ouste","outre","ouvert","ouverte","ouverts","o|","o","p","paf","pan","par","parce","parfois","parle","parlent","parler","parmi","parseme","partant","particulier","particulire","particulirement","pas","pass","pendant","pense","permet","personne","peu","peut","peuvent","peux","pff","pfft","pfut","pif","pire","plein","plouf","plus","plusieurs","plutt","possessif","possessifs","possible","possibles","pouah","pour","pourquoi","pourrais","pourrait","pouvait","prealable","precisement","premier","premire","premirement","pres","probable","probante","procedant","proche","prs","psitt","pu","puis","puisque","pur","pure","q","qu","quand","quant","quant--soi","quanta","quarante","quatorze","quatre","quatre-vingt","quatrime","quatrimement","que","quel","quelconque","quelle","quelles","quelqu'un","quelque","quelques","quels","qui","quiconque","quinze","quoi","quoique","r","rare","rarement","rares","relative","relativement","remarquable","rend","rendre","restant","reste","restent","restrictif","retour","revoici","revoil","rien","s","sa","sacrebleu","sait","sans","sapristi","sauf","se","sein","seize","selon","semblable","semblaient","semble","semblent","sent","sept","septime","sera","seraient","serait","seront","ses","seul","seule","seulement","si","sien","sienne","siennes","siens","sinon","six","sixime","soi","soi-mme","soit","soixante","son","sont","sous","souvent","specifique","specifiques","speculatif","stop","strictement","subtiles","suffisant","suffisante","suffit","suis","suit","suivant","suivante","suivantes","suivants","suivre","superpose","sur","surtout","t","ta","tac","tant","tardive","te","tel","telle","tellement","telles","tels","tenant","tend","tenir","tente","tes","tic","tien","tienne","tiennes","tiens","toc","toi","toi-mme","ton","touchant","toujours","tous","tout","toute","toutefois","toutes","treize","trente","tres","trois","troisime","troisimement","trop","trs","tsoin","tsouin","tu","t","u","un","une","unes","uniformement","unique","uniques","uns","v","va","vais","vas","vers","via","vif","vifs","vingt","vivat","vive","vives","vlan","voici","voil","vont","vos","votre","vous","vous-mmes","vu","v","vtre","vtres","w","x","y","z","zut","","","a","s","taient","tais","tait","tant","t","tre",""]
STOPWORDS_MODIF = STOPWORDS.union(set(['alternate']))

#Load the spacy french lemmatizer and the count(document number currently being process)
def setCounterAndNlp():
    global count 
    count = 0
    global nlp
    nlp = spacy.load('fr_core_news_sm',disable=['parser', 'tagger', 'ner'])
    nlp.max_length = 1000000

#Lemmatize english text
def english_lemmatize_stemming(text):
    return WordNetLemmatizer().lemmatize(text, pos='v')

"""
This function preprocess the text buy tokenizing words and finding lemma for the right language.
Inputs:
        - Decoded texts
Outputs:
        - List of list of lemmatized words for each texts
Time: 
        - 1-2 Hours
"""
def preprocess(text):
    
    global count
    global nlp
    
    # Lowercase text and remove punctations
    result = []
    text = re.sub(r'[^\w]', ' ', text).lower()
    language = ""
    
    #Find the two first languages and proportions using the first 100000 words at most.(faster)
    if len(text)>100000:
        
        isReliable, textBytesFound, details = cld2.detect(text[:100000])
        lang1,lang2 = details[0], details[1]
    else:
        isReliable, textBytesFound, details = cld2.detect(text)
        lang1,lang2 = details[0], details[1]
        
    # Check if it is a mix of both french and english
    if ((lang1.language_name == "FRENCH" and lang2.language_name == "ENGLISH") or (lang1.language_name == "ENGLISH" and lang2.language_name == "FRENCH")) and lang1.percent < 80:
        
        language = "ENGLISH_FRENCH_MIX"
        
    # French or English   
    elif lang1.language_name == "FRENCH" or lang1.language_name == "ENGLISH":
        
        language = lang1.language_name
        
    # English is the default language        
    else:
        language = "ENGLISH"
    
    
    
    # ENGLISH-FRENCH document. Because of that mix, each world need to be analyzed independently.
    
    if language == "ENGLISH_FRENCH_MIX":  # ENGLISH_FRENCH
        
        print(count,". ",language," Size: ",len(text), " Percent: ", lang1.percent,"",lang1.language_code,", ", lang2.percent,"",lang2.language_code)
        
        # For every token, lemmatize it in french and english, find if one or both lemmatizers 
        # removed some part and if yes, only keep the lemmatization of the most dominant language
        for token in gensim.utils.simple_preprocess(text):
        
            if (token not in STOPWORDS_MODIF and len(token) > 3) and (token not in FRENCH_STOPWORDS):
                            
                eng_lemma = english_lemmatize_stemming(token) # english lemma
                unaccented_string = unidecode.unidecode(token)
                fr_lemma = nlp(token)[0].lemma_ # french lemma
                            
                if eng_lemma != token and fr_lemma != token:
                    
                    if lang1.language_name == "FRENCH" and lang1.percent > lang2.percent:
                        result.append(fr_lemma)
                    elif lang1.language_name == "ENGLISH" and lang1.percent > lang2.percent:
                        result.append(eng_lemma)
                        
                elif fr_lemma != token:
            
                    result.append(fr_lemma)
                
                #english is the default lemma
                else:
                    result.append(eng_lemma)
        
    
    # FRENCH document. Because scapy cannot cope with large text exceeding the RAM of the computer.
    # Each text, if too big has been divided in smaller part scapy can handle.
                                                                  
    elif language =='FRENCH': 
       
        print(count,". ",language," Size: ",len(text))
        len_text = len(text)
        max_per_pass = nlp.max_length
        
        # Document is too long. Lemmatization is done in multiple steps.
        if(len_text > max_per_pass):
            
            nb_pass = math.floor(len_text/max_per_pass)+1
            remainder = len_text % max_per_pass
            print("Nb of pass: ",nb_pass,", text: ",len_text," characters")
            for i in range(nb_pass):
                
                if i<nb_pass-1:
                    doc = nlp(text[i*max_per_pass : (i+1)*max_per_pass-1])
                    pass_result = [token.lemma_ for token in doc if token.text not in FRENCH_STOPWORDS if len(token.text)>3 if bool(token.text.strip()) if not bool(re.search(r'\d', token.text))]
                    result.extend(pass_result)
                    print("Pass ",str(i)," From ", i*max_per_pass, " to ",(i+1)*max_per_pass-1,", result is now of size: ", len(result))
                else:
                    doc = nlp(text[i*max_per_pass: (i*max_per_pass) + remainder-1])
                    pass_result = [token.lemma_ for token in doc if token.text not in FRENCH_STOPWORDS if len(token.text)>3 if bool(token.text.strip()) if not bool(re.search(r'\d', token.text))]
                    result.extend(pass_result)
                    print("Pass ",str(i)," From ", i*max_per_pass, " to ",(i*max_per_pass) + remainder-1,", result is now of size: ", len(result))
                    
        # Document is short. Lemmatization is done in one go.     
        else:
            doc = nlp(text) 
            result = [token.lemma_ for token in doc if token.text not in FRENCH_STOPWORDS if len(token.text)>3 if bool(token.text.strip()) if not bool(re.search(r'\d', token.text))]
    
    
    # ENGLISH or OTHER document
    elif language == "ENGLISH":                    
        
        print(count,". ",language," Size: ",len(text))
        # For each token, lemmatize it using nltk.
        for token in gensim.utils.simple_preprocess(text):
        
            if (token not in STOPWORDS_MODIF and len(token) > 3):
                lemma = english_lemmatize_stemming(token)
                result.append(lemma)

    
    
    count = count+1
    return result


#Lemmatize the whole dataframe    
setCounterAndNlp()
process_textDf = textDf['text'].map(preprocess)
```

                                    3. DICTIONARY / BAG OF WORDS

```python

# Get a dictionary of the words present in all lemmatized texts.
dictionary = gensim.corpora.Dictionary(process_textDf)

#Remove words than are not present a lot and words that are too present. 
# 20 = nb of min occurence of a word in all texts.
# 0.4 = word need to be present in less than 40 % of the texts.
dictionary.filter_extremes(no_below= 20, no_above=0.4, keep_n=500000)

# Get o bag of words corpus.
bow_corpus = [dictionary.doc2bow(doc) for doc in process_textDf]
```

4.3 Optimal number of topics search

```python
                              # 2.4 Modelling and best number of topic search

"""In order to find the optimal number of topics, multiple models with increasing number of topics are fitted. 
   Coherence values for each model is then used to infer the optimal number of topics. 
   
   Inputs:
       - Dictionary         
       - Bag of words corpus
       - Lemmatized texts.
       - Max topic number
       - Starting number of topic
       - Increasing topic step
       
   Outputs:
       - The list of all models built
       - Coherence values for all models.
       
   Time:
       - 3-4 hours
"""

def compute_coherence_values(dictionary, corpus, texts, limit, start=10, step=2):
  
    coherence_values = []
    model_list = []
    
    for num_topics in range(start, limit, step):
        print("Starting model ",num_topics)
        model = gensim.models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=2)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
        print("Finished")
        print(coherencemodel.get_coherence())

    return model_list, coherence_values


#Fit models and find coherences for model with 20-29 topics.
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=bow_corpus, texts=process_textDf, start=20, limit=30, step=1)

# Print the coherence values for the increasing number of topics.
limit=30; start=20; step=1;
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
for model in model_list:
    if model.num_topics == 29:
        lda_model = model
        for idx, topic in model.print_topics(-1):
            print('Topic: {} \nWords: {}'.format(idx, topic))
```

                                     3. TRAINING TOPIC/COSINE DATAFRAME

```python
"""
This function creates for each text, a score of appartenance for each topic betweeen 0 and 1.
The sum of all scores for one text is egal to 1.
This allows to get a representation of the most dominant topics of each texts.
Inputs: 
        - Final LDA model
        - Bag of words corpus
        - Lemmatized texts
Outputs: 
        - A large dataframe with scores of topics for each texts.
        
Time: 
        - 1-10 minute
"""

def topicsForSentences(ldamodel,corpus, texts):
    
    topic_series = process_textDf
    topic_series.name = "text"
    topic_df = topic_series.to_frame()
    for i in range(ldamodel.num_topics):
        topicName = "Topic " + str(i)
        topic_df[topicName] = 0.0
    print("inside")
    
    # Get main topics in each document
    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        
        # Get the dominant topics and percentile contribututions 
        for j, (topic_num, prop_topic) in enumerate(row):
            
            topicName = "Topic " + str(topic_num)
            topic_df.at[i,topicName] = prop_topic

    return topic_df


topic_df = topicsForSentences(model_list[0],bow_corpus,process_textDf)
#topic_df = topicsForSentences(ldamallet,bow_corpus,process_textDf)
```

```python
"""
This final step allows for a large dataframe with only topics, Node ids, and cossine similarity scores between 
two nodes text informations to be created.
"""

# Load the training Dataframe and only keep the node ids and link

finalDf = pd.read_csv('Data/training.txt',delimiter=" ",header=None)
finalDf.columns = ["Node1","Node2","Link"]
finalDf["Cos_similarity"] = 0.0

#lda_model = model_list[9] # Most performing model with 29 topics
lda_model = ldamallet # Most performing model with 29 topics
# Add topic columns for each node.

for i in range(lda_model.num_topics):
    topic_name = "1_Topic "+ str(i)
    finalDf[topic_name] = 0.0

for i in range(lda_model.num_topics):
    topic_name = "2_Topic "+ str(i)
    finalDf[topic_name] = 0.0 
    
# For every text, add the topics scores and cosine similarity score between words.    
for index,row in finalDf.iterrows():
    
    if (index == 50000):
        print (50000)
        
    idx1 = int(row["Node1"])
    idx2 = int(row["Node2"])

    cos_sim = gensim.matutils.cossim(lda_model[bow_corpus[idx1]], lda_model[bow_corpus[idx2]])
    finalDf.at[index,"Cos_similarity"] = cos_sim
    
    n_1_name = "1_Topic "
    n_2_name = "2_Topic "
    
    for i,(v1,v2) in enumerate(zip(topic_df.iloc[idx1][1:lda_model.num_topics+1],topic_df.iloc[idx2][1:lda_model.num_topics+1])):
        
        n1_topic_name = n_1_name + str(i)
        n2_topic_name = n_2_name + str(i)
        
        finalDf.at[index, n1_topic_name] = v1
        finalDf.at[index, n2_topic_name] = v2   
```
